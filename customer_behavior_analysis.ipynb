```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')

df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
df
```

```python
# create a colunm age group
labels = ['Young Adult', 'Adult', 'middle-eged', 'Senior']

df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)

df[['age', 'age_group']].sample(10)
```

```python
# create column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 6,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
import pymysql
import pandas as pd
from sqlalchemy import create_engine

# Database credentials
db_name = "customer_behavior"
db_host = "localhost" # or your MySQL server IP
db_username = "root"
db_password = "root1234"
db_port = 3306 # Default MySQL port

# Create a SQLAlchemy engine for connecting
engine = create_engine(f"mysql+pymysql://{db_username}:{db_password}@{db_host}:{db_port}/{db_name}")

try:
    conn = engine.connect()
    print("Connection to MySQL successful!")
except Exception as e:
    print(f"Error connecting to MySQL: {e}")
```

```python
# Write DataFrame to MySQL.
df.to_sql("customer", conn, if_exists="replace", index=False)
print("DataFrame successfully written to MySQL.")
```
